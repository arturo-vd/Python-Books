#Multi function calculator

```python
import sympy
from sympy import symbols
from sympy.solvers import solve
import math
from fractions import Fraction

#----- 1. Basic Arithmetic ---

def basic_arithmetic():
    print("\n --- Basic Arithmetic ---")
    try:
        n1 = float(input("Enter first number: "))
        op = input("Enter operation (+, -, *, /): ")
        n2 = float(input("Enter second number: "))
        if op =='+':
           print(f"Result: {n1 + n2}")
        elif op == '-':
             print(f"Result: {n1 -n2}")
        elif op =='*':
             print({f"Result: {n1 * n2}"})
        elif op == '/':
            if n2 == 0:
                print("Error: Division by zero.")
            else:
              print(f"Result: {n1 / n2}")

        else:
            print("Invalid pperator.")

    except ValueError:
        print("Invalid input. Please enter valid numbers.")

# ----- 2. Algebra: Solve for x ---

def solve_for_x():
  print("\n--- Solver for x ---")
  print("Instructions: Enter equation set to zero (e.g., 2*x -4).")
  print("Always use '*' for multiplication (e.g., 2*x, not 2x).")

  eq_input = input("0 = ")
  x = symbols('x')

  try:
    # solve() returns a list of solutions
    solution = solve(eq_input, x)
    if len(solution) > 0:
       print(f"x = {solution[0]}")
       #Check if there are multiple solutions (like for quafratics)
       if len(solution) > 1:
          print(f"Other solutions: {solution[1:]}")
    else:
        print("No solution found.")

  except:
       print("Error: Could not solve. Check your syntax.")

#----- 3. Proportions ---

def solve_proportion():
    print("\n--- Solve Proportions ---")
    print("Format: n1 / d1 = n2 / d2")
    print("Enter the three known values. Type 'x' for the unknown.")

    n1_in = input("n1 = ")
    d1_in = input("d1 = ")
    n2_in = input("n2 = ")
    d2_in = input("d2 = ")

    try:
        # Logic: If n2 is 'x', then n2 = (n1 * d2) / d1
        if n2_in.lower() == 'x':
            n1, d1, d2 = float(n1_in), float(d1_in), float(d2_in)
            print(f"n2 = {(n1 * d2) / d1}")

        # Logic: If d2 is 'x', then d2 = (n2 * d1) / n1
        elif d2_in.lower() == 'x':
            n1,d1,n2 = float(n1_in), float(d1_in), float(n2_in)
            print(f"d2 = {(n2 * d1) / n1}")

        # Logic: If n1 is 'x' , then n1 = (n2 * d1)  / d2
        elif n1_in.lower() == 'x':
            n2, d1, d2 = float(n2_in), float(d1_in), float(d2_in)
            print(f"n1 = {(n2 * d1) / d2}")
        # Logic: If d1 is 'x', then d1 = (n1 * d2) / n2
        elif d1_in.lower() == 'x':
            n1, d2, n2 = float(n1_in), float(d2_in), float(n2_in)
            print(f"d1 = {(n1 * d2) / n2}")
        else:
            print("Error: You must type 'x' for exactly one variable.")
    except ValueError:
         print("Error: Please enter valid numbers for the knowns.")

# ---- 4. Factor Square Roots ---

def factor_square_roots():
    print(f"\n --- Factor Square Roots ---")
    try:
        n= int(input("Enter a number to factor the square root of: "))
        upper_limit = int(n**0.5) + 1
        max_square = 1

        # Find the largest perfect square factor
        for i in range(1, upper_limit):
            if n % (i**2) == 0:
                max_square = i**2

        square_root_part = int(max_square**0.5)
        remaining_part = int(n / max_square)

        if max_square == 1:
            print(f"Result: sqrt({n}) (Cannot be simplified further)")
        elif remaining_part == 1:
            print(f"Result: {square_root_part}")
        else:
            print(f"Result: sqrt({n}) = sqrt({square_root_part}) * {remaining_part}")
    except ValueError:
        print("Error: Please enter a valid number.")

#----- 5. Conversions --

def decimal_to_fractions_percent():
    print("\n--- Decimal -> Fraction & Percent ---")
    try:
        digits = input("Enter a decimal number: ")
        # Calculate decimal placer to determine denominator
        if '.' in digits:
          decimal_places = len(digits.split('.')[1])
        else:
          decimal_places = 0

        n = float(digits)
        denominator = 10**decimal_places
        numerator = int(n * denominator)
        percent = n * 100

        print(f"Decimal: {n}")
        print(f"Fraction: {numerator}/{denominator}")
        print(f"Percent: {percent}%")

    except ValueError:
        print("Error: Please enter a valid decimal number.")


def fraction_to_decimal_percent():
    print("\n--- Fraction -> Decimal & Percent ---")
    try:

        num = float(input("Numerator:"))
        den = float(input("Denominator: "))
        if den == 0:
            print("Error: Division by zero.")
        else:
            dec = num /den
            print(f"Decimal: {dec}")
            print(f"Percent: {dec * 100}%")

    except ValueError:
        print("Error: Please enter valid numbers for the numerator and denominator.")

def percent_to_decimal_fraction():
    print("\n--- Percent -> Decimal & Fraction ---")
    try:
        p = float(input("Enter percent (without %): "))
        dec = p / 100

        # This converts the decimal into a simplified fraction
        frac = Fraction(dec).limit_denominator()

        print(f"Decimal: {dec}")
        print(f"Fraction: {frac}")

    except ValueError:
        print("Invalid input.")

# ----- MAIN MENU ---

def main_menu():
    while True:
        print("\n" + "="*30)
        print(" Multi-Function Calculator Menu")
        print("\n" + "="*30)
        print("1. Basic Arithmetic (+ - * /)")
        print("2. Solve for x (Algebra)")
        print("3. Solve Proportions")
        print("4. Factor Square Roots")
        print("5. Convert Decimals")
        print("6. Convert Fractions")
        print("7. Convert Percents")
        print("8. Exit")

        choice = input("Enter your choice (1-8): ")

        if choice == '1':
            basic_arithmetic()
        elif choice == '2':
            solve_for_x()
        elif choice == '3':
            solve_proportion()
        elif choice == '4':
            factor_square_root()
        elif choice == '5':
            decimal_to_fraction_percent()
        elif choice == '6':
            fraction_to_decimal_percent()
        elif choice == '7':
            percent_to_decimal_fraction()
        elif choice == '8':
            print("Goodbye! Thanks for computing.")
            break
        else:
            print("Invalid choice, please try again.")

main_menu()



```
